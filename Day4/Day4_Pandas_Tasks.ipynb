## 5. Dataset Information

```python
# Shape of the dataset (rows, columns)
print("Shape:", df.shape)
```

## 6. Quick Insights

```python
# Top scorer by Total marks
top_scorer = df.loc[df['Total'].idxmax()]
print("Top Scorer:\n", top_scorer)
```

```python
# Average marks per subject
print("Average Math:", df['Math'].mean())
print("Average Science:", df['Science'].mean())
print("Average English:", df['English'].mean())
```

```python
# Students who scored above 80% percentage
df[df['Percentage'] > 80]
```
