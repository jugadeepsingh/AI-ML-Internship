# Day 5 – Data Cleaning Tasks
**Topics:** Handle missing values, remove duplicates, and understand dataset statistics.

**Expected Outcome:** Clean dataset prepared.

## 1. Import Libraries & Load the Raw Dataset
We use `student_scores_raw.csv` — a version of the Day 4 dataset with some missing values and duplicate rows intentionally introduced, to practice cleaning.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/loq/Downloads/student_scores_raw.csv")
print("Dataset loaded successfully!")
df
```

## 2. Understand Dataset Statistics (Before Cleaning)

```python
# Shape of the dataset
print("Shape:", df.shape)
```

```python
# Dataset info (data types, non-null counts)
df.info()
```

```python
# Statistical summary
df.describe()
```

## 3. Handle Missing Values

```python
# Check missing values per column
df.isnull().sum()
```

```python
# Percentage of missing values per column
(df.isnull().sum() / len(df)) * 100
```

```python
# Rows containing missing values
df[df.isnull().any(axis=1)]
```

```python
# Option 1: Fill missing numeric values with column mean
df['Math'] = df['Math'].fillna(df['Math'].mean())
df['Science'] = df['Science'].fillna(df['Science'].mean())
df['English'] = df['English'].fillna(df['English'].mean())

# Option 2: Fill missing categorical/text values with a placeholder
df['Name'] = df['Name'].fillna("Unknown")

print("Missing values after filling:\n", df.isnull().sum())
```

## 4. Remove Duplicates

```python
# Check number of duplicate rows
print("Duplicate rows before cleaning:", df.duplicated().sum())
```

```python
# View the duplicate rows
df[df.duplicated()]
```

```python
# Drop duplicate rows
df = df.drop_duplicates().reset_index(drop=True)
print("Duplicate rows after cleaning:", df.duplicated().sum())
print("New shape:", df.shape)
```

## 5. Recalculate Derived Columns
Since Math/Science/English values changed (missing values filled), recompute Total and Percentage for consistency.

```python
df['Total'] = df['Math'] + df['Science'] + df['English']
df['Percentage'] = round(df['Total'] / 3, 2)
df.head()
```

## 6. Dataset Statistics (After Cleaning)

```python
# Confirm no missing values remain
print("Missing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
# Final statistical summary
df.describe()
```

```python
df.info()
```

## 7. Save the Cleaned Dataset

```python
df.to_csv("student_scores_clean.csv", index=False)
print("Clean dataset saved as student_scores_clean.csv")
```

- Handled missing values: filled numeric columns with mean, filled text column with a placeholder
- Identified and removed duplicate rows using `duplicated()` and `drop_duplicates()`
- Recalculated derived columns and verified the dataset was clean
